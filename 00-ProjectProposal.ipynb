## Research Question

Which factors most strongly predict customer ratings in food delivery orders, and does delivery time have a larger impact than food preparation time?



## Background and Prior Work

Online food delivery or on-demand food delivery, has become an essential part of consumers’ everyday eating behaviors. The consumers can search for and order food from various restaurants on digital platforms, then couriers deliver the food to the consumers’ doorstep. This allows consumers to reduce time costs for food purchases by eliminating the physical effort of commuting and waiting in line at food establishments. According to a paper “Delivery time reliability in on-demand food delivery: Heterogeneity from attribution effects,” they emphasize the significance of on demand food delivery by stating how “The global ODFD revenue has increased by more than 60 percent, from 254.7 billion US dollars in 2020 to 424.5 billion US dollars in 2024 (Statista, 2024)” (Ma et al.). Since the online food delivery market has been growing significantly, the competition among the delivery platforms has increased as well. Therefore, these platforms have to run their operations more efficiently to remain profitable without hurting customer satisfaction. So these platforms have to understand how consumers think and behave (e.g., how they value delivery time, reliability, delays, and service promises), then use this information to build better decision-making models (e.g., pricing, delivery-time promises, dispatching, and capacity allocation). These platforms can no longer simply deliver and provide service, but make effective decisions while being aware of what makes things profitable for them. From there, our group was interested in how factors of food delivery like pricing, delivery time, food quality, and more can contribute to customer satisfaction since these reviews on food delivery platforms take many factors into consideration besides food quality.

Looking through online Git projects, we were able to find a few repositories focusing on Food Delivery Times through the usage of prediction models. The repo that we observed the most is delivery_time_prediction, with its creator cleaning a publicly provided dataset on Kaggle, and optimizing it for the prediction model. After exploring a few possible relationships and hypotheses, the variables that ended up contributing the most to food delivery time were: age of the delivery driver, rating of the delivery driver, and the distance between restaurant and delivery location. These findings align with our research question, and poses more inquiry whether or not these results are similar across different scenarios and datasets. 

https://github.com/MisterAare/delivery_time_prediction 

We also were able to find a published research paper very similarly focused on our project topic. In this study, the authors, Yootetch el al. (2022) surveyed 411 experienced users of food delivery apps and analyzed factors such as menu variety, food quality, delivery quality, and the perceived value of the app. They used structural equation modeling to analyze the survey data and found that perceived app value had the strongest influence on customer satisfaction, followed by delivery quality. Factors such as menu variety and food quality actually had weaker effects on ratings. Their results suggest that customers are more affected by how reliable, efficient, and convenient the delivery experience is rather than food preparation factors alone. Based on these results, we expect to observe similar patterns in our project, with delivery-related factors having a stronger influence on customer satisfaction.
