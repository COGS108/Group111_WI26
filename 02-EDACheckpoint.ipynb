## Background and Prior Work

&nbsp;&nbsp;&nbsp;&nbsp;Online food delivery or on-demand food delivery, has become an essential part of consumers’ everyday eating behaviors. Digital platforms allow customers to browse restaurant options, place orders, and receive food at their doorstep, significantly reducing the time and effort associated with in-person dining or takeout. According to a paper “Delivery time reliability in on-demand food delivery: Heterogeneity from attribution effects,” they emphasize the significance of on demand food delivery by stating how “The global ODFD revenue has increased by more than 60 percent, from 254.7 billion US dollars in 2020 to 424.5 billion US dollars in 2024 (Statista, 2024)” (Ma et al.). As the market expands, competition among delivery platforms has increased as well. To remain profitable while maintaining customer satisfaction, platforms run their operations more efficiently by understanding how consumers think and behave (e.g., how they value delivery time, reliability, delays, and service promises), then use this information to build better decision-making models (e.g., pricing, delivery-time promises, dispatching, and capacity allocation). These platforms no longer simply deliver and provide service, but make effective decisions while being aware of what makes things profitable for them. <br>
&nbsp;&nbsp;&nbsp;&nbsp;Additionally, the research from the previous paper suggests delivery cost and expected delivery time negatively influence ODFD usage, although customer sensitivity varies across contexts. For example, consumers appear less sensitive to delays during peak hours or long-distance orders, likely because expectations adjust to anticipated congestion. Early deliveries generally increase satisfaction, while late deliveries reduce usage, though the negative effect of delays may be weaker for long-distance orders. Based on these results, we saw how the cost of food delivery and expected delivery time are the factors that have a lot of impact on the ratings and satisfaction of their experience with online food delivery. But this study emphasizes how the factors that contribute to the customers’ satisfaction to expected delivery time varies based on customers like location, day of the week, time of the week, distance, what they were doing, etc. Therefore, we want to include other factors that can contribute to customer satisfaction with delivery time which will be reflected in their ratings. The factors we considered that are included in our dataset are time of the food order and what day of the week did the customer order as these factors will contribute to the customers’ service expectations which contributes to their ratings of their food delivery orders. From there, our group narrowed down our interests to factors of food delivery like pricing, delivery time, and more that can contribute to customer satisfaction since these reviews on food delivery platforms take many factors into consideration besides food quality. <br>
&nbsp;&nbsp;&nbsp;&nbsp;After researching more information about food delivery, we came across a relevant peer-reviewed study that provides theoretical grounding for our research question is “Determinants of Customer Satisfaction via Online Food Delivery Applications”, by Chanin Yoopetch, Pisit Siriphan, and Suthawan Chirapanda (2022). In this study, the authors surveyed 411 users of food delivery apps and used structural equation modeling to analyze how menu variety, food quality, delivery quality, and perceived app value influence customer satisfaction. Their findings showed that perceived value had the strongest effect on restaurant satisfaction, followed by delivery quality, while menu variety and food quality had comparatively weaker direct effects. Overall, the results indicate that delivery performance (including speed, reliability, and overall efficiency) has a stronger and more consistent influence on customer satisfaction than food preparation-related factors alone. These findings suggest that customer may be more sensitive to delivery issues than preparation delays and provide theoretical support for our research question. However, applying these findings requires consideration of alternative perspectives and potential threats to validity. Although this prior research highlights the importance of delivery-related factors, customer satisfaction is multidimensional, meaning factors such as preparation time, cuisine type, order cost, and ordering context (e.g. day of the week) may still influence outcomes. While cuisine-related or menu based characteristics are included in our analysis, they are not a primary focus, as the referenced study found menu-related factors to have comparatively smaller effects on satisfaction outcomes. Instead, our project emphasizes operational variables more directly tied to delivery performance.<br>

## Exploratory Data Analysis

In this section, we explore the structure and key patterns within the processed Grubhub dataset. The goal of this exploratory data analysis (EDA) is to understand the distribution of important variables and identify potential relationships between delivery performance and customer ratings.

We begin by examining the distribution of delivery times to better understand how long orders typically take to arrive. Understanding this distribution helps determine whether delivery times are normally distributed or skewed, which can inform later statistical modeling decisions.

Next, we investigate potential relationships between delivery time and customer ratings. If slower deliveries negatively affect customer satisfaction, we would expect to observe a negative relationship between delivery time and rating values.

Throughout this section, visualizations and correlation tests are used to evaluate patterns in the data before moving on to more formal statistical analysis.

### Checking for Normality and Linearity

Before applying statistical models, it is useful to examine whether the variables follow approximately normal distributions and whether relationships between variables appear linear.

To assess this, we first visualize the distribution of delivery times using a histogram and kernel density estimate (KDE). This helps determine whether delivery times are normally distributed or exhibit skewness.

We also create binned versions of delivery time to simplify comparisons and help visualize potential trends in customer ratings across different delivery time ranges. These visualizations provide an initial understanding of the structure of the data before applying correlation tests or regression models.

```python
# Grubhub Plots
plt.figure(figsize=(8,5))

sns.histplot(
    data=grubhub,
    x='delivery_time',
    bins=20,            
    kde=True
)

plt.title("Distribution of Delivery Time")# Create bins
grubhub['delivery_bin'] = pd.cut(grubhub['delivery_time'], bins=6)

plt.figure(figsize=(10,5));
```

#### Observations

The histogram shows the distribution of delivery times across the dataset. Most deliveries appear to occur within a moderate time range, with the highest density concentrated around roughly 25–40 minutes.

The distribution appears slightly right-skewed, meaning that while most deliveries occur within a typical window, there are some orders that take substantially longer. These longer delivery times form the right tail of the distribution.

This skewness suggests that delivery performance varies across orders, possibly due to factors such as distance, restaurant preparation time, or traffic conditions. Because the distribution is not perfectly normal, non-parametric correlation tests (such as Spearman correlation) may be appropriate when examining relationships between delivery time and customer ratings in later analysis.

### Relationship Between Delivery Time and Customer Ratings

To explore whether delivery speed influences customer satisfaction, we visualize the relationship between delivery time and customer ratings.

Because ratings are discrete values ranging from 1 to 5, many observations overlap on the same horizontal lines. To make the distribution of points clearer, we add a small amount of random noise (jitter) to the rating values. This allows overlapping points to become visible while preserving the overall structure of the data.

We then plot delivery time against ratings and color observations by state. Regression lines are included for each state to help visualize potential trends between delivery time and customer ratings across different geographic regions.

```python
# Grubhub Plots
# Add small jitter to rating if discrete (1-5)
grubhub['rating_jitter'] = grubhub['rating'] + np.random.uniform(-0.1, 0.1, size=len(grubhub))

# Scatter plot with regression lines per state
sns.lmplot(
    data=grubhub,
    x='delivery_time',
    y='rating_jitter',
    hue='state',           # updated column name
    height=6,
    aspect=1.5,
    scatter_kws={'alpha':0.6},
    line_kws={'linewidth':2}
)

plt.title("Delivery Time vs Rating with Regression Lines by State")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Rating")
plt.show()
```

#### Observations

The scatterplot shows a large concentration of ratings between 3 and 5 across all delivery times, suggesting that most orders receive relatively positive customer feedback.

The regression lines indicate that the relationship between delivery time and ratings is relatively weak and varies slightly across states. Some states show a mild positive trend while others show a slight negative trend, suggesting that delivery time alone may not strongly determine customer satisfaction.

This initial visualization suggests that other factors, such as order cost, restaurant quality, or preparation time, may play a larger role in influencing customer ratings. In the following sections, we use statistical tests such as Spearman correlation and regression models to more formally evaluate these relationships.

### Delivery Time vs Rating by State

To better understand whether delivery speed affects customer satisfaction, we examine the relationship between delivery time and customer ratings for several states in the dataset.

Because ratings are discrete values ranging from 1 to 5, many observations overlap when plotted directly. To make the distribution of points clearer, we add a small amount of random noise (jitter) to the rating values. This helps reveal overlapping observations without meaningfully changing the data.

We then create separate scatterplots for selected states (NY, NJ, CA, IL, and MA). Each plot shows delivery time on the x-axis and customer rating on the y-axis. A regression line is added to each plot to visualize the overall trend between delivery time and ratings within each state.

```python
# Add small jitter to ratings (1–5)
grubhub['rating_jitter'] = grubhub['rating'] + np.random.uniform(-0.1, 0.1, size=len(grubhub))

# List of states you want separate plots for
states = ['NY','NJ','CA','IL','MA']

plt.figure(figsize=(18, 10))

for i, state in enumerate(states, 1):
    plt.subplot(2, 3, i)  # 2 rows, 3 columns
    subset = grubhub[grubhub['state'] == state]
    
    # Scatter plot
    sns.scatterplot(
        data=subset,
        x='delivery_time',
        y='rating_jitter',
        alpha=0.6,
        color='skyblue'
    )
    
    # Regression line
    sns.regplot(
        data=subset,
        x='delivery_time',
        y='rating_jitter',
        scatter=False,
        color='darkred',
        line_kws={'linewidth':2}
    )
    
    plt.title(f"State: {state}")
    plt.xlabel("Delivery Time (minutes)")
    plt.ylabel("Rating")

plt.tight_layout()
plt.show()
```

#### Observations

Across all states, customer ratings appear heavily concentrated between 3 and 5, suggesting that most orders receive relatively positive feedback regardless of delivery time.

The regression lines show slightly different patterns across states. Some states show a weak positive relationship between delivery time and ratings, while others show a slight negative trend. However, the slopes appear relatively small, indicating that delivery time alone may not be a strong predictor of customer satisfaction.

This preliminary visualization suggests that additional factors—such as restaurant quality, order price, or service fees—may play a more significant role in determining customer ratings. ### Comparing Ratings Across States: Mann–Whitney U Test

To examine whether customer ratings differ between states, we conduct a Mann–Whitney U test comparing ratings in New York and New Jersey. This non-parametric test is used because it does not assume that the rating distributions are normally distributed.

The test evaluates whether the distributions of ratings between the two states are statistically different.

```python
#Testing Mann Whitney U-test
group1 = grubhub[grubhub['state'] == 'NY']['rating']
group2 = grubhub[grubhub['state'] == 'NJ']['rating']

stat, p = mannwhitneyu(group1, group2, alternative='two-sided')

print(f"Mann-Whitney U statistic: {stat:.2f}")
print(f"P-value: {p:.4e}")

if p < 0.05:
    print("Result: Significant difference between NY and NJ ratings")
else:
    print("Result: No significant difference between NY and NJ ratings")
```

```python
# Checking for Effect Size
group1 = grubhub[grubhub['state']=='NY']['rating']
group2 = grubhub[grubhub['state']=='NJ']['rating']

stat, p = mannwhitneyu(group1, group2, alternative='two-sided')

n1 = len(group1)
n2 = len(group2)
N = n1 + n2

mean_U = n1*n2 / 2
std_U = np.sqrt(n1*n2*(n1+n2+1)/12)
Z = (stat - mean_U)/std_U

r = Z / np.sqrt(N)

print(f"Effect size r: {r:.3f}")
```

#### Interpretation

The Mann–Whitney U test results indicate a statistically significant difference between ratings in New York and New Jersey (p < 0.05). This suggests that the distribution of customer ratings differs between the two states.

However, the calculated effect size (r ≈ 0.058) is very small. This indicates that while the difference is statistically significant, the magnitude of the difference between the two states is minimal.

In practice, this suggests that although ratings may differ slightly between New York and New Jersey, the difference is unlikely to be practically meaningful.

This observation is consistent with the earlier correlation and statistical analyses, which also indicated very weak relationships between delivery time and ratings across platforms.

## Final Conclusion

This analysis examined whether delivery time meaningfully influences customer ratings across multiple food delivery platforms. The central hypothesis proposed that longer delivery times would negatively affect customer satisfaction, either by lowering ratings or reducing the likelihood that customers leave ratings altogether. This expectation was based on the idea that delivery speed is an important component of perceived service quality in food delivery services. In particular, we predicted that customers in regions with higher service expectations—such as the West Coast and Northeast—might respond more negatively to longer delivery times.

To evaluate this hypothesis, we analyzed datasets from three food delivery platforms: FoodHub, Grubhub, and DoorDash. Each dataset provided information on operational variables such as delivery time, preparation time, order cost, and customer ratings. The analysis began with exploratory data analysis (EDA) to understand the structure of the data and identify potential patterns. Visualizations such as pairplots, scatterplots, and hexbin density plots were used to examine how ratings varied across delivery times and other operational factors. These visual tools allowed us to observe general trends and identify whether longer delivery times appeared to correspond with lower ratings.

Correlation heatmaps were then used to quantify the relationships between ratings and key variables such as delivery time, preparation time, and order cost. Across all three datasets, the correlation values between delivery time and ratings were extremely small and close to zero. This indicated that there was little to no linear relationship between these variables. In other words, increases in delivery time did not appear to correspond with meaningful decreases in customer ratings.

To further investigate the relationship between delivery time and ratings, we conducted Spearman rank correlation tests. Spearman correlation was selected because ratings are ordinal in nature and may not follow a strictly linear relationship with delivery time. The results of these tests were consistent across platforms: while some correlations were statistically significant due to the large number of observations in the datasets, the correlation coefficients themselves were extremely small. This indicates that the practical effect of delivery time on ratings is negligible.

The visualizations further supported this conclusion. Hexbin density plots showed that the majority of customer ratings cluster between 4 and 5 across nearly all delivery times. Even when delivery times increase, ratings remain concentrated in this high range. This pattern suggests that customers continue to provide relatively positive ratings regardless of moderate variations in delivery duration. The absence of a clear downward trend in ratings as delivery time increases provides strong visual evidence that delivery time alone does not strongly influence rating behavior.

Taken together, the results across FoodHub, Grubhub, and DoorDash consistently suggest that the initial hypothesis is not strongly supported. **Although longer delivery times might intuitively seem likely to reduce customer satisfaction, the data indicate that ratings remain relatively stable even when delivery times vary**. This suggests that other factors may be more important in shaping customer evaluations of their orders.

One possible explanation is that customer ratings reflect the overall service experience rather than a single operational factor. Elements such as food quality, order accuracy, restaurant reputation, packaging, or customer expectations may have a stronger impact on ratings than delivery speed alone. Additionally, customers may anticipate some variability in delivery times, especially during busy periods, and therefore may not penalize longer waits unless they exceed expectations by a substantial margin.

There are also several limitations to consider in this analysis. First, the datasets do not capture all aspects of the delivery experience. Variables such as restaurant popularity, order complexity, driver availability, weather conditions, or traffic patterns could influence both delivery times and customer satisfaction. Second, ratings themselves may contain bias, as customers who leave reviews may not represent the entire population of customers. Finally, differences in dataset structure and platform-specific behaviors may influence how ratings are distributed across platforms.

Future research could expand on this analysis by incorporating additional variables that better capture the full customer experience. For example, including information about restaurant type, order size, promotional discounts, or customer location could provide deeper insights into what factors most strongly influence customer satisfaction. Machine learning models could also be used to explore whether combinations of variables predict ratings more effectively than any single factor alone.

Overall, the findings from this study suggest that delivery time, while operationally important, is not a strong standalone predictor of customer ratings in food delivery services. The consistent results across multiple platforms reinforce the conclusion that customer satisfaction in food delivery is influenced by a broader set of factors beyond delivery speed alone.
