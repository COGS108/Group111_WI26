## Background and Prior Work

&nbsp;&nbsp;&nbsp;&nbsp;Online food delivery or on-demand food delivery, has become an essential part of consumers’ everyday eating behaviors. Digital platforms allow customers to browse restaurant options, place orders, and receive food at their doorstep, significantly reducing the time and effort associated with in-person dining or takeout. According to a paper “Delivery time reliability in on-demand food delivery: Heterogeneity from attribution effects,” they emphasize the significance of on demand food delivery by stating how “The global ODFD revenue has increased by more than 60 percent, from 254.7 billion US dollars in 2020 to 424.5 billion US dollars in 2024 (Statista, 2024)” (Ma et al.). As the market expands, competition among delivery platforms has increased as well. To remain profitable while maintaining customer satisfaction, platforms run their operations more efficiently by understanding how consumers think and behave (e.g., how they value delivery time, reliability, delays, and service promises), then use this information to build better decision-making models (e.g., pricing, delivery-time promises, dispatching, and capacity allocation). These platforms no longer simply deliver and provide service, but make effective decisions while being aware of what makes things profitable for them. <br>
&nbsp;&nbsp;&nbsp;&nbsp;Additionally, the research from the previous paper suggests delivery cost and expected delivery time negatively influence ODFD usage, although customer sensitivity varies across contexts. For example, consumers appear less sensitive to delays during peak hours or long-distance orders, likely because expectations adjust to anticipated congestion. Early deliveries generally increase satisfaction, while late deliveries reduce usage, though the negative effect of delays may be weaker for long-distance orders. Based on these results, we saw how the cost of food delivery and expected delivery time are the factors that have a lot of impact on the ratings and satisfaction of their experience with online food delivery. But this study emphasizes how the factors that contribute to the customers’ satisfaction to expected delivery time varies based on customers like location, day of the week, time of the week, distance, what they were doing, etc. Therefore, we want to include other factors that can contribute to customer satisfaction with delivery time which will be reflected in their ratings. The factors we considered that are included in our dataset are time of the food order and what day of the week did the customer order as these factors will contribute to the customers’ service expectations which contributes to their ratings of their food delivery orders. From there, our group narrowed down our interests to factors of food delivery like pricing, delivery time, and more that can contribute to customer satisfaction since these reviews on food delivery platforms take many factors into consideration besides food quality. <br>
&nbsp;&nbsp;&nbsp;&nbsp;After researching more information about food delivery, we came across a relevant peer-reviewed study that provides theoretical grounding for our research question is “Determinants of Customer Satisfaction via Online Food Delivery Applications”, by Chanin Yoopetch, Pisit Siriphan, and Suthawan Chirapanda (2022). In this study, the authors surveyed 411 users of food delivery apps and used structural equation modeling to analyze how menu variety, food quality, delivery quality, and perceived app value influence customer satisfaction. Their findings showed that perceived value had the strongest effect on restaurant satisfaction, followed by delivery quality, while menu variety and food quality had comparatively weaker direct effects. Overall, the results indicate that delivery performance (including speed, reliability, and overall efficiency) has a stronger and more consistent influence on customer satisfaction than food preparation-related factors alone. These findings suggest that customer may be more sensitive to delivery issues than preparation delays and provide theoretical support for our research question. However, applying these findings requires consideration of alternative perspectives and potential threats to validity. Although this prior research highlights the importance of delivery-related factors, customer satisfaction is multidimensional, meaning factors such as preparation time, cuisine type, order cost, and ordering context (e.g. day of the week) may still influence outcomes. While cuisine-related or menu based characteristics are included in our analysis, they are not a primary focus, as the referenced study found menu-related factors to have comparatively smaller effects on satisfaction outcomes. ### FoodHub EDA

To better understand how operational factors may influence customer ratings, we analyze the FoodHub dataset as a **comparison case**. This dataset contains information on order cost, preparation time, delivery time, and customer ratings for food delivery orders.

We begin with descriptive statistics and visualizations to understand the distribution of the variables. In particular, we examine how ratings are distributed and whether operational variables such as delivery time, preparation time, and order cost appear to relate to customer satisfaction.

Because the rating variable is discrete and skewed toward higher values, we later rely on statistical tests that are more appropriate for this type of data when evaluating potential relationships.

```python
foodhub.describe()
```

To begin the FoodHub analysis, we first examined basic descriptive statistics for the dataset. The dataset contains 1,162 observations of food delivery orders. On average, the order cost is about $16.76, the average preparation time is about 27 minutes, and the average delivery time is about 24 minutes. The average customer rating is 4.34, indicating that most orders receive relatively high ratings.

Looking at the quartiles, the median rating is 5, and both the 75th percentile and maximum rating are also 5. This suggests that ratings are heavily concentrated at the upper end of the scale, meaning customers tend to give high ratings. Because the rating variable only takes a few values (primarily 3, 4, and 5), the distribution is not continuous and is skewed toward higher ratings.

To better visualize this, we plotted a histogram of the rating distribution, which shows that most ratings fall at 4 or 5, with relatively few lower ratings. This skewed distribution helps explain why it may be difficult to detect strong relationships between ratings and operational variables such as delivery time or preparation time.

```python
sns.histplot(foodhub['rating'], bins=5)
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

Ratings are very similar for weekday and weekend orders. Both groups show a median near 5 and a similar spread between 4 and 5, suggesting that the day of the week does not meaningfully affect customer ratings.

```python
sns.boxplot(x='day_type', y='rating', data=foodhub)

plt.title("Ratings by Day Type")
plt.xlabel("Day Type")
plt.ylabel("Rating")

plt.show()
```

Weekday ratings average 4.31 and weekend ratings average 4.36, so the difference is very small. A two-sample t-test gives a t-statistic of −1.05 and a p-value of 0.295. Since the p-value is greater than 0.05, there is no statistically significant difference between weekday and weekend ratings. This suggests that day type does not meaningfully affect customer ratings in the FoodHub dataset.

```python
print(foodhub.groupby("day_type")["rating"].mean())
weekday = foodhub[foodhub['day_type'] == 'Weekday']['rating']
weekend = foodhub[foodhub['day_type'] == 'Weekend']['rating']
t_stat, p_value = stats.ttest_ind(weekday, weekend)
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
sns.heatmap(foodhub[['rating','delivery_time','prep_time','order_cost']].corr(),
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Matrix")
plt.show()
```

The correlation matrix shows that ratings have almost no linear relationship with delivery time, preparation time, or order cost. All correlation values are extremely close to zero (−0.01 to 0.03), indicating that changes in these variables do not meaningfully correspond with changes in customer ratings. To further examine whether these factors collectively influence ratings, we fit an Ordinary Least Squares (OLS) regression model using delivery time, preparation time, order cost, and day type as predictors of rating.

The OLS regression results show that none of the variables significantly predict ratings. All predictors have p-values greater than 0.05, and the model has an R² of 0.002, meaning it explains almost none of the variation in ratings. This further suggests that delivery time, preparation time, order cost, and whether the order occurs on a weekend do not meaningfully influence customer ratings in the FoodHub dataset.
