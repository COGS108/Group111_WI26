## Research Question

Using statistical inference and regression analysis, which factors (delivery time, food cost, time of order, and day of the week) significantly influence customer ratings in FoodHub restaurant food delivery orders in New York, and is the effect of delivery time statistically larger than that of food preparation time?

## Background and Prior Work

Online food delivery or on-demand food delivery, has become an essential part of consumers’ everyday eating behaviors. The consumers can search for and order food from various restaurants on digital platforms, then couriers deliver the food to the consumers’ doorstep. This allows consumers to reduce time costs for food purchases by eliminating the physical effort of commuting and waiting in line at food establishments. According to a paper “Delivery time reliability in on-demand food delivery: Heterogeneity from attribution effects,” they emphasize the significance of on demand food delivery by stating how “The global ODFD revenue has increased by more than 60 percent, from 254.7 billion US dollars in 2020 to 424.5 billion US dollars in 2024 (Statista, 2024)” (Ma et al.). Since the online food delivery market has been growing significantly, the competition among the delivery platforms has increased as well. Therefore, these platforms have to run their operations more efficiently to remain profitable without hurting customer satisfaction. So these platforms have to understand how consumers think and behave (e.g., how they value delivery time, reliability, delays, and service promises), then use this information to build better decision-making models (e.g., pricing, delivery-time promises, dispatching, and capacity allocation). These platforms no longer simply deliver and provide service, but make effective decisions while being aware of what makes things profitable for them. 

Additionally, it was observed from this paper that delivery cost and expected delivery time does negatively impact the use of on demand food delivery (ODFD), but the time sensitivity (consumers’ satisfaction in response to delivery time) varies on the consumers. Consumers are generally less sensitive to delivery time during peak hours and of long-distance orders, likely due to them adjusting their service expectations to be more lenient with these circumstances. Early deliveries tend to increase overall usage of ODFD which leads to the idea that people generally like receiving orders earlier than expected, however, there are consumers who dislike early delivery, potentially because they aren’t ready to receive or haven’t arrived at a designated spot for pick up. While late deliveries are widely seen as bad service and decreases ODFD usage. But the late deliveries for long distance orders are less negatively effective because consumers may be more lenient, expecting uncertainty and forgiving of delays. The paper developed and tested their own hypotheses as well. For instance, there was no significant effect between blaming the late delivery to riders and consumer utility of ODFD for Hypothesis 1 (Consumers will exhibit lower levels of dissatisfaction (lower reliability sensitivity) if they tend to attribute the delivery unreliability to delivery riders). Hypothesis 2 is supported (if customers blame the restaurant, they feel more negative about the late delivery). Hypothesis 3 is not supported (if customers blame the platform, they feel less negative). 

Based on these results, we saw how the cost of food delivery and expected delivery time are the factors that have a lot of impact on the ratings and satisfaction of their experience with online food delivery. But this study emphasizes how the factors that contribute to the customers’ satisfaction to expected delivery time varies based on customers like location, day of the week, time of the week, distance, what they were doing, etc. Therefore, we want to include other factors that can contribute to customer satisfaction with delivery time which will be reflected in their ratings. The factors we considered that are included in our dataset are time of the food order and what day of the week did the customer order as these factors will contribute to the customers’ service expectations which contributes to their ratings of their food delivery orders. From there, our group narrowed down our interests to factors of food delivery like pricing, delivery time, and more that can contribute to customer satisfaction since these reviews on food delivery platforms take many factors into consideration besides food quality. Along with these factors, we also took into consideration the hypotheses the paper tested when developing our own hypothesis. 


After researching more information about food delivery, we came across a relevant peer-reviewed study that provides theoretical grounding for our research question is “Determinants of Customer Satisfaction via Online Food Delivery Applications”, by Chanin Yoopetch, Pisit Siriphan, and Suthawan Chirapanda (2022). In this study, the authors surveyed 411 users of food delivery apps and used structural equation modeling to analyze how menu variety, food quality, delivery quality, and perceived app value influence customer satisfaction. Their findings showed that perceived value had the strongest effect on restaurant satisfaction, followed by delivery quality, while menu variety and food quality had comparatively weaker direct effects. 


Overall, the results indicate that delivery performance (including speed, reliability, and overall efficiency) has a stronger and more consistent influence on customer satisfaction than food preparation-related factors alone. These findings suggest that customer may be more sensitive to delivery issues than preparation delays and provide theoretical support for our research question. However, applying these findings requires consideration of alternative perspectives and potential threats to validity. Although this prior research highlights the importance of delivery-related factors, customer satisfaction is multidimensional, meaning factors such as preparation time, cuisine type, order cost, and ordering context (e.g. day of the week) may still influence outcomes. While cuisine-related or menu based characteristics are included in our analysis, they are not a primary focus, as the referenced study found menu-related factors to have comparatively smaller effects on satisfaction outcomes. Instead, our project emphasizes operational variables more directly tied to delivery performance.


Looking through online Git projects, we were able to find a few repositories focusing on Food Delivery Times through the usage of prediction models. The repo that we observed the most is delivery_time_prediction, with its creator cleaning a publicly provided dataset on Kaggle, and optimizing it for the prediction model. After exploring a few possible relationships and hypotheses, the variables that ended up contributing the most to food delivery time were: age of the delivery driver, rating of the delivery driver, and the distance between restaurant and delivery location. These findings align with our research question, and poses more inquiry whether or not these results are similar across different scenarios and datasets.


Therefore, for our group project, we aim to narrow our attention to the factors that have been proven to show potential impact on customer satisfaction such as food delivery time. In our project, we will define food preparation time as the time (in minutes) taken by the restaurant to prepare the food which is calculated by taking the difference between the timestamps of the restaurant's order confirmation and the delivery person's pick-up confirmation. While defining delivery time as the time taken by the delivery person to deliver the food package which is calculated by taking the difference between the timestamps of the delivery person's pick-up confirmation and drop-off information.
