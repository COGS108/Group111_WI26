## Background and Prior Work

Online food delivery or on-demand food delivery, has become an essential part of consumers’ everyday eating behaviors. The consumers can search for and order food from various restaurants on digital platforms, then couriers deliver the food to the consumers’ doorstep. This allows consumers to reduce time costs for food purchases by eliminating the physical effort of commuting and waiting in line at food establishments. According to a paper “Delivery time reliability in on-demand food delivery: Heterogeneity from attribution effects,” they emphasize the significance of on demand food delivery by stating how “The global ODFD revenue has increased by more than 60 percent, from 254.7 billion US dollars in 2020 to 424.5 billion US dollars in 2024 (Statista, 2024)” (Ma et al.). Since the online food delivery market has been growing significantly, the competition among the delivery platforms has increased as well. Therefore, these platforms have to run their operations more efficiently to remain profitable without hurting customer satisfaction. So these platforms have to understand how consumers think and behave (e.g., how they value delivery time, reliability, delays, and service promises), then use this information to build better decision-making models (e.g., pricing, delivery-time promises, dispatching, and capacity allocation). These platforms no longer simply deliver and provide service, but make effective decisions while being aware of what makes things profitable for them. 
Additionally, it was observed from this paper that delivery cost and expected delivery time does negatively impact the use of on demand food delivery (ODFD), but the time sensitivity (consumers’ satisfaction in response to delivery time) varies on the consumers. Consumers are generally less sensitive to delivery time during peak hours and of long-distance orders, likely due to them adjusting their service expectations to be more lenient with these circumstances. Early deliveries tend to increase overall usage of ODFD which leads to the idea that people generally like receiving orders earlier than expected, however, there are consumers who dislike early delivery, potentially because they aren’t ready to receive or haven’t arrived at a designated spot for pick up. While late deliveries are widely seen as bad service and decreases ODFD usage. But the late deliveries for long distance orders are less negatively effective because consumers may be more lenient, expecting uncertainty and forgiving of delays. 
Based on these results, we saw how the cost of food delivery and expected delivery time are the factors that have a lot of impact on the ratings and satisfaction of their experience with online food delivery. But this study emphasizes how the factors that contribute to the customers’ satisfaction to expected delivery time varies based on customers like location, day of the week, time of the week, distance, what they were doing, etc. Therefore, we want to include other factors that can contribute to customer satisfaction with delivery time which will be reflected in their ratings. The factors we considered that are included in our dataset are time of the food order and what day of the week did the customer order as these factors will contribute to the customers’ service expectations which contributes to their ratings of their food delivery orders. From there, our group narrowed down our interests to factors of food delivery like pricing, delivery time, and more that can contribute to customer satisfaction since these reviews on food delivery platforms take many factors into consideration besides food quality.

Their findings showed that perceived value had the strongest effect on restaurant satisfaction, followed by delivery quality, while menu variety and food quality had comparatively weaker direct effects. 
Overall, the results indicate that delivery performance (including speed, reliability, and overall efficiency) has a stronger and more consistent influence on customer satisfaction than food preparation-related factors alone. These findings suggest that customer may be more sensitive to delivery issues than preparation delays and provide theoretical support for our research question. However, applying these findings requires consideration of alternative perspectives and potential threats to validity. Instead, our project emphasizes operational variables more directly tied to delivery performance.

### Comparing Ratings Across States: Mann–Whitney U Test

To examine whether customer ratings differ between states, we conduct a Mann–Whitney U test comparing ratings in New York and New Jersey. This non-parametric test is used because it does not assume that the rating distributions are normally distributed.

The test evaluates whether the distributions of ratings between the two states are statistically different.

```python
# Creating subset dataframes
NY_ratings = grubhub[grubhub['state'] == 'NY']['rating']
NJ_ratings = grubhub[grubhub['state'] == 'NJ']['rating']

# Full code can be found in eda_functions
mwu_test(NY_ratings, NJ_ratings, "NY", "NJ")
```

#### Interpretation

The Mann–Whitney U test results indicate a statistically significant difference between ratings in New York and New Jersey (p < 0.05). This suggests that the distribution of customer ratings differs between the two states.

However, the calculated effect size (r ≈ 0.058) is very small. This indicates that while the difference is statistically significant, the magnitude of the difference between the two states is minimal.

In practice, this suggests that although ratings may differ slightly between New York and New Jersey, the difference is unlikely to be practically meaningful.
