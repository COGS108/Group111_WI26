### Grubhub Dataset

* Dataset #1
  * Name: "Grubhub Restaurant Data"
  * Link: https://www.kaggle.com/datasets/polartech/grubhub-restaurant-data?resource=download
  * Observations: 8088
  * Variables: 27
  * Description of the variables:
    - "searched_state": The state in which the order was placed in
    - "delivery_time": The time in minutes that the order took to arrive from the restaurant to the ordering address
    - "review_rating" The rating given by the customer on a 0-5 scale

```python
grubhub = pd.read_csv('data/00-raw/grubhub.csv')
grubhub.head()
```

The dataset above is very robust with 27 columns and over 8000 rows (wow!), but to relate back to our research question, we need to specifically focus on state, delivery time, and review rating. We can drop the other columns that aren't of any interest.

```python
# Drop irrelevant columns
grubhub_drop_columns = ['searched_zipcode', 'searched_city', 'searched_lat', 'searched_lng', 'searched_address', 'searched_metro', 
                        'is_gh', 'latitude', 'longitude', 'distance', 'cuisines', 'delivery_fee', 'service_fee', 'loc_name', 'loc_number',
                        'url', 'address', 'phone', 'RunDate', 'restaurant_tags', 'delivery_fee_raw', 'delivery_time_raw', 'review_count', 'delivery_type']

grubhub = grubhub.drop(columns = grubhub_drop_columns)

# Simplify column names to match the other dataset
rename_cols = {'searched_state': 'state',
                'delivery_time': 'delivery_time',
                'review_rating': 'rating'}

grubhub = grubhub.rename(columns = rename_cols)

# Checking for null entries. There are none
print(grubhub.isnull().sum())
print(grubhub.shape)
```

Delivery_time can contain a lot of variability! Especially with so many datapoints, so it's best to filter out any outliers. After that, we are done cleaning the Grubhub dataset.

```python
# Removing any potential outliers for delivery time
Q1 = grubhub['delivery_time'].quantile(0.25)
Q3 = grubhub['delivery_time'].quantile(0.75)
IQR = Q3 - Q1

grubhub = grubhub[(grubhub['delivery_time'] >= Q1 - 1.5*IQR) &
                  (grubhub['delivery_time'] <= Q3 + 1.5*IQR)]

grubhub.head()
```

* Dataset #3

        * "day_of_the_week": The day of the week in which the order was placed
        * "rating": The rating given by the customer on a 0-5 scale
        * "food_preparation_time": The time in minutes it took for the food to be prepared in the restaurant
        * "delivery_time": The time in minutes it took for the food to be delivered from the Restaurant to the order address


```python
foodhub = pd.read_csv('data/00-raw/food_order.csv')
foodhub
```

We can keep the columns of interest. We can also see that under the rating column, there exists an entry "Not given". Analysis can't be performed on ratings that don't exist, so we can treat these entries as null and remove them from the dataframe.

```python
# Dropping columns
drop_cols = ['order_id', 'customer_id', 'restaurant_name', 'cuisine_type']
foodhub = foodhub.drop(columns = drop_cols)

# Dropping "Not given" values
foodhub = foodhub[foodhub['rating'] != 'Not given']
foodhub['rating'] = foodhub['rating'].astype(float)
```

After that, we also have to make sure there are no other null entries.

```python
# Counting up all null entries per column; there are none
print(foodhub.isnull().sum())

# Show how many rows and columns we're left with after cleaning
print(foodhub.shape)
```
