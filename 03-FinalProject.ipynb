### Grubhub Dataset

* Dataset #1
  * Name: "Grubhub Restaurant Data"
  * Link: https://www.kaggle.com/datasets/polartech/grubhub-restaurant-data?resource=download
  * Observations: 8088
  * Variables: 27
  * Description of the variables:
    - "searched_state": The state in which the order was placed in
    - "delivery_time": The time in minutes that the order took to arrive from the restaurant to the ordering address
    - "review_rating" The rating given by the customer on a 0-5 scale

```python
grubhub = pd.read_csv('data/00-raw/grubhub.csv')
grubhub.head()
```

The dataset above is very robust with 27 columns and over 8000 rows (wow!), but to relate back to our research question, we need to specifically focus on state, delivery time, and review rating among other things. We can drop the other columns that aren't of any interest.

```python
# Drop irrelevant columns
grubhub_drop_columns = ['searched_zipcode', 'searched_lat', 'searched_lng', 'searched_address', 'searched_metro', 
                        'is_gh', 'latitude', 'longitude', 'cuisines', 'loc_name', 'loc_number', 'url', 'address', 'phone', 
                        'RunDate', 'restaurant_tags', 'delivery_fee_raw', 'delivery_time_raw']

grubhub = grubhub.drop(columns = grubhub_drop_columns)

# Simplify column names to match the other dataset
rename_cols = {'searched_state': 'state',
                'delivery_time': 'delivery_time',
                'review_rating': 'rating'}

grubhub = grubhub.rename(columns = rename_cols)

# Checking for null entries. There are none
print(grubhub.isnull().sum())
print(grubhub.shape)
```

Delivery_time can contain a lot of variability! Especially with so many datapoints, so it's best to filter out any outliers. After that, we are done cleaning the Grubhub dataset.

```python
# Removing any potential outliers for delivery time
Q1 = grubhub['delivery_time'].quantile(0.25)
Q3 = grubhub['delivery_time'].quantile(0.75)
IQR = Q3 - Q1

grubhub = grubhub[(grubhub['delivery_time'] >= Q1 - 1.5*IQR) &
                  (grubhub['delivery_time'] <= Q3 + 1.5*IQR)]

grubhub.head()
```

```python
group1 = grubhub[grubhub['state'] == 'NY']['rating']
group2 = grubhub[grubhub['state'] == 'NJ']['rating']

stat, p = mannwhitneyu(group1, group2, alternative='two-sided')

print(f"Mann-Whitney U statistic: {stat:.2f}")
print(f"P-value: {p:.4e}")

if p < 0.05:
    print("Result: Significant difference between NY and NJ ratings")
else:
    print("Result: No significant difference between NY and NJ ratings")
```

```python
group1 = grubhub[grubhub['state']=='NY']['rating']
group2 = grubhub[grubhub['state']=='NJ']['rating']

stat, p = mannwhitneyu(group1, group2, alternative='two-sided')

n1 = len(group1)
n2 = len(group2)
N = n1 + n2

mean_U = n1*n2 / 2
std_U = np.sqrt(n1*n2*(n1+n2+1)/12)
Z = (stat - mean_U)/std_U

r = Z / np.sqrt(N)

print(f"Effect size r: {r:.3f}")
```

To better understand how operational factors may influence customer ratings, we analyze the FoodHub dataset as a **comparison case**. This dataset contains information on order cost, preparation time, delivery time, and customer ratings for food delivery orders.

### Cross-Platform Correlation Analysis of Delivery Time and Ratings

To determine whether delivery time meaningfully influences customer ratings across food delivery platforms, we compare the correlation results from the FoodHub, Grubhub, and DoorDash datasets. Across all three datasets, the relationships between delivery time and customer ratings appear extremely weak.

The correlation heatmaps show that ratings have almost no linear relationship with delivery time, preparation time, or order cost. All Pearson correlation values are very close to zero, suggesting that variations in these operational variables do not correspond strongly with changes in customer ratings.

To further evaluate the relationship between delivery time and ratings, we computed the Spearman rank correlation. The results are consistent across platforms: the correlation coefficients are very close to zero, indicating only negligible associations. Although some tests produce statistically significant p-values due to the large sample sizes, the effect sizes remain extremely small.

Overall, the consistency of these findings across multiple datasets suggests that delivery time alone is not a strong predictor of customer ratings. Other factors—such as food quality, restaurant reputation, or overall service experience—may play a more substantial role in determining how customers evaluate their orders.
