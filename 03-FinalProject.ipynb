## Background and Prior Work

Online food delivery or on-demand food delivery, has become an essential part of consumers’ everyday eating behaviors. The consumers can search for and order food from various restaurants on digital platforms, then couriers deliver the food to the consumers’ doorstep. This allows consumers to reduce time costs for food purchases by eliminating the physical effort of commuting and waiting in line at food establishments. According to a paper “Delivery time reliability in on-demand food delivery: Heterogeneity from attribution effects,” they emphasize the significance of on demand food delivery by stating how “The global ODFD revenue has increased by more than 60 percent, from 254.7 billion US dollars in 2020 to 424.5 billion US dollars in 2024 (Statista, 2024)” (Ma et al.). Since the online food delivery market has been growing significantly, the competition among the delivery platforms has increased as well. Therefore, these platforms have to run their operations more efficiently to remain profitable without hurting customer satisfaction. So these platforms have to understand how consumers think and behave (e.g., how they value delivery time, reliability, delays, and service promises), then use this information to build better decision-making models (e.g., pricing, delivery-time promises, dispatching, and capacity allocation). These platforms no longer simply deliver and provide service, but make effective decisions while being aware of what makes things profitable for them. Additionally, it was observed from this paper that delivery cost and expected delivery time does negatively impact the use of on demand food delivery (ODFD), but the time sensitivity (consumers’ satisfaction in response to delivery time) varies on the consumers. Consumers are generally less sensitive to delivery time during peak hours and of long-distance orders, likely due to them adjusting their service expectations to be more lenient with these circumstances. Early deliveries tend to increase overall usage of ODFD which leads to the idea that people generally like receiving orders earlier than expected, however, there are consumers who dislike early delivery, potentially because they aren’t ready to receive or haven’t arrived at a designated spot for pick up. ### Checking for Normality and Linearity

Before applying statistical models, we need to check whether the variables are approximately normal and whether their relationships appear linear.

To assess this, we first visualize the distribution of delivery times using a histogram and kernel density estimate (KDE). From this, we can see its shape and where a majority of the data is clustered.

```python
plt.figure(figsize=(8,5));

sns.histplot(
    data=grubhub,
    x='delivery_time',
    bins=20,            
    kde=True
)

plt.title("Distribution of Delivery Time")# Create bins
grubhub['delivery_bin'] = pd.cut(grubhub['delivery_time'], bins=6)

plt.figure(figsize=(10,5));
```

The histogram shows that the distribution of delivery times roughly span between 20 to 70 minutes, with a high concentration of orders being delivered within 25-40 minutes. The shape is slightly right skewed, meaning that while most deliveries occur within a typical window, some can take much longer, causing the right tail of the distribution.

This skewness suggests that delivery performance varies across orders, possibly due to dexternal factors such as distance, restaurant preparation time, or traffic conditions. Non-parametric correlation tests (such as Spearman correlation) may be appropriate when examining relationships between delivery time and customer ratings in later analysis because of the lack of normality.

### Relationship Between Delivery Time and Customer Ratings

To explore whether delivery speed influences customer satisfaction, we can visualize the relationship between delivery time and customer ratings on a scatterplot.

Because ratings are discrete values ranging from 1 to 5, many observations overlap on the same horizontal lines. To make the distribution of points clearer, we add a small amount of random noise (jitter) to the rating values, allowing overlapping points to be visible. We then plot delivery time against ratings and color observations by state. Regression lines are included for each state to help visualize potential trends between delivery time and customer ratings across different locations.

```python
# Add small jitter to rating to better visualize the plotted data
grubhub['rating_jitter'] = grubhub['rating'] + np.random.uniform(-0.1, 0.1, size=len(grubhub))

# To view the full code, refer to eda_functions.py
state_lmplot(grubhub)
```

Here, we can see the scatterplot shows a large concentration of ratings between 3 and 5 across all delivery times, suggesting that most orders receive relatively positive customer feedback. The regression lines indicate that the relationship between delivery time and ratings is relatively weak and varies slightly across states. Some states show a mild positive trend while others show a slight negative trend, suggesting that delivery time alone may not strongly determine customer satisfaction.

For further analysis, we can break this scatterplot into subplots separated by state.

```python
# Full code in eda_functions.py
state_scatterplot(grubhub)
```

Now that we can see everything more clearly, some states show a weak positive relationship between delivery time and ratings, while others show a slight negative trend. The positive trend would indicate that longer delivery times yield a higher rating, which intuitively doesn't make much sense, and we also have to recognize that the data forms a "cloud" cluster, making the linear regression line inconclusive in the current state that the data is presented in. This and the weak slopes indicate that delivery time alone may not be a strong predictor of customer satisfaction.

This preliminary visualization suggests that additional factors such as restaurant quality, order price, or service fees, may play a more significant role in determining customer ratings. To investigate this further, we next examine correlation statistics and regression models to quantify these relationships.

### Correlation Matrix of Key Variables

To better understand the relationships between the main variables in the dataset, we compute a correlation matrix for ratings, distance, delivery fees, delivery time, service fees, and review counts. The heatmap visualizes the strength and direction of linear relationships between these variables.

### Multivariable OLS Regression: Predicting Customer Ratings

To examine how multiple delivery-related factors jointly relate to customer ratings, we can attempt to use a multivariable Ordinary Least Squares (OLS) regression model. The model predicts ratings using delivery time, delivery fees, service fees, delivery distance, and state-level fixed effects. While our data may not be parametric, we can examine the results and see how it runs. 


```python
formula = 'rating ~ delivery_time + delivery_fee + service_fee + distance + C(state)'

model = ols(formula, data=grubhub_test_2).fit()

print(model.summary())
```

The regression results indicate that several delivery-related variables have statistically significant relationships with customer ratings, although the overall explanatory power of the model is relatively low (R² ≈ 0.075).

Delivery time shows a small positive coefficient, suggesting that longer delivery times are weakly associated with slightly higher ratings when controlling for other factors. However, the magnitude of this relationship is very small.

Both delivery fees and service fees show negative coefficients, indicating that higher fees are associated with slightly lower customer ratings. Delivery distance also has a small negative relationship with ratings.

State-level effects are statistically significant, suggesting that average ratings differ across states even after controlling for delivery characteristics.

Overall, while several variables are statistically significant, the low R² indicates that these delivery-related factors explain only a small portion of the variation in customer ratings. This means that other factors that were not present in the Grubhub dataset like customer expectations may play a larger role in evaluating food delivery satisfaction.

### Comparing Ratings Across States: Mann–Whitney U Test

From the OLS regression, we learn that there might be a significant difference in ratings between different US states, so we can examine this by conducting a Mann–Whitney U test comparing ratings in New York and New Jersey. We pivot to this non-parametric test as it does not assume normality, which may have been a potential limitation for our data to begin with.

The test evaluates whether the distributions of ratings between the two states are statistically different.

```python
# Creating subset dataframes
NY_ratings = grubhub[grubhub['state'] == 'NY']['rating']
NJ_ratings = grubhub[grubhub['state'] == 'NJ']['rating']

# Full code can be found in eda_functions
mwu_test(NY_ratings, NJ_ratings, "NY", "NJ", "ratings")
```

#### Interpretation

The Mann–Whitney U test results indicate a statistically significant difference between ratings in New York and New Jersey (p < 0.05). This suggests that the distribution of customer ratings differs between the two states.

However, the calculated effect size (r ≈ 0.058) is very small. This indicates that while the difference is statistically significant, the magnitude of the difference between the two states is minimal.

In practice, this suggests that although ratings may differ slightly between New York and New Jersey, the difference is unlikely to be practically meaningful.

With this lack of linearity and the skewness, we've learned that an OLS regression model won't be applicable in this case. Instead, we'll default to calculating the Spearman Rank correlation for preparation time, order cost, and delivery time.

```python
d_corr, d_p = spearmanr(foodhub['delivery_time'], foodhub['rating'])
p_corr, p_p = spearmanr(foodhub['prep_time'], foodhub['rating'])
c_corr, c_p = spearmanr(foodhub['order_cost'], foodhub['rating'])

print("Spearman Correlation Results")
print("----------------------------")
print(f"Delivery Time vs Rating: correlation = {d_corr:.3f}, p-value = {d_p:.3f}")
print(f"Prep Time vs Rating:     correlation = {p_corr:.3f}, p-value = {p_p:.3f}")
print(f"Order Cost vs Rating:    correlation = {c_corr:.3f}, p-value = {c_p:.3f}")
```

All Spearman correlation coefficients are extremely close to zero, and none are statistically significant (p > 0.05). This indicates that there is no meaningful monotonic relationship between these factors and customer ratings in the FoodHub dataset.
